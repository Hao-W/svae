```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import math
from data import *
```

```python
dims = (100, 100, 20)
balls = [(10, 10, 4, 0.5), (20, 80, 8, 0.5), (50, 50, 10, 0.5)]
frames, R = gen_frames(dims, balls)
```

```python
gs = gridspec.GridSpec(2, 10)
gs.update(left=0.0 , bottom=0.0, right=1.0, top=1.0, wspace=0.0, hspace=0)
fig = plt.figure(figsize=(40, 8))
for i in range(frames.shape[2]):
    frame = frames[:,:,i]
    ax = fig.add_subplot(gs[int(i/10), int(i%10)])
    ax.imshow(frame, cmap='Greys')
    ax.set_xticks([])
    ax.set_yticks([])
```
