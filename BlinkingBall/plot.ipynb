```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import math
from data import *
```

```python
pixels = 64
radi = 10
K = 3
dims = (pixels, pixels, 1)
NUM_SEQS = 10000
```

```python
DATA = np.zeros((NUM_SEQS, pixels, pixels))
MU = np.zeros((NUM_SEQS, K, 2))
for i in range(NUM_SEQS):
    mu = np.random.uniform(radi, pixels - radi, (K, 2))
    balls = [(int(mu[0,0]), int(mu[0,1]), radi, 1),
             (int(mu[1,0]), int(mu[1,1]), radi, 1),
             (int(mu[2,0]), int(mu[2,1]), radi, 1)]
    frames, R = gen_frames(dims, balls)
    DATA[i] = frames[:, :, 0]
    MU[i] = mu
    if i % 1000 == 0:
        print(i)
```

```python
path = "../../BBs"
np.save(path + '/ob', DATA)
np.save(path + '/mu', MU)
```

```python
gs = gridspec.GridSpec(1, 10)
gs.update(left=0.0 , bottom=0.0, right=1.0, top=1.0, wspace=0.0, hspace=0)
fig = plt.figure(figsize=(40, 8))
for i in range(frames.shape[2]):
    frame = frames[:,:,i]
    ax = fig.add_subplot(gs[int(i/10), int(i%10)])
    ax.imshow(frame, alpha=0.8)
    ax.set_xticks([])
    ax.set_yticks([])
```
