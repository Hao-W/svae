```python
pixels = 20
T = 50
num_seqs = 100
Ys = np.zeros((num_seqs, pixels, T))
for n in range(num_seqs):
    Y = np.zeros((pixels, T))
    pos = np.random.randint(pixels)
    direction = np.random.choice(np.array([-1,1]))
    for t in range(T):
        if pos == 0 and direction == -1:
            direction = 1
        elif pos == (pixels-1) and direction == 1:
            direction = -1
        pos += direction
        Y[pos, t] = 1
#     plt.imshow(Y, cmap='gray', vmin=0.0, vmax=1.0)
    Ys[n] = Y
```

```python
T = 50
theta1 = np.pi / 32
theta2 = np.pi / 16

damp = 1.0
A1 = np.array([[np.cos(theta1), -np.sin(theta1)],[np.sin(theta1), np.cos(theta1)]])
A2 = np.array([[np.cos(theta2), np.sin(theta2)],[-np.sin(theta2), np.cos(theta2)]])
A = np.zeros((4,4))
A[:2, :2] = A1
A[2:, 2:] = A2

num_seqs = 200
Xs = np.zeros((num_seqs, T, 4))
for n in range(num_seqs):
    X = np.zeros((T, 4))
    x = Uniform(torch.ones(4) * (-1),torch.ones(4) * 1).sample().data.numpy()[:, None]
    X[0] = x[:, 0]
    for t in range(1, T, 1):
        noise = mvn.rvs(mean=np.zeros(4), cov=np.eye(4) * (0.02 ** 2))
        xnew = np.dot(A, x) + noise
        x = xnew
        X[t] = x[:, 0]
    Xs[n] = X
```

```python
plt.plot(X[:, 0])
plt.plot(X[:, 1])
plt.plot(X[:, 2])
plt.plot(X[:, 3])
```

```python
np.save('signals/sequences', Xs)
```
