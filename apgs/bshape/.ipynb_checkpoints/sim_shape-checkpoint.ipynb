```python
%matplotlib inline
from apgs.bshape.sim_bshape import Sim_BShape
timesteps = 10
num_objects = 4 
frame_size = 50 
object_size = 6 
offset = 2
dv = 0.2
chunk_size = 1000
simulator = Sim_BShape(timesteps, num_objects, frame_size, object_size, offset, dv, chunk_size)
```

```python
simulator.viz_data()
```

```python
%run sim_bshape.py --chunk_size=2000
```

```python
objects = simulator.generate_objects()
```

```python
import matplotlib.pyplot as plt
for k in range(len(objects)):
    plt.imshow(objects[k], cmap='gray')
    plt.show()
```

```python
import numpy as np
import torch
ball = objects[0]
for i in range(3):
    if i == 0:
        ball3d = torch.cat((ball[None,:,:], torch.zeros(10, 10)[None,:,:], torch.zeros(10,10)[None,:,:]), 0)
    elif i == 1:
        ball3d = torch.cat((torch.zeros(10, 10)[None,:,:], ball[None,:,:], torch.zeros(10,10)[None,:,:]), 0)
    else:
        ball3d = torch.cat((torch.zeros(10, 10)[None,:,:], torch.zeros(10,10)[None,:,:], ball[None,:,:]), 0)
    plt.imshow(ball3d.permute(1,2,0), vmin=0, vmax=1.0)
    plt.show()
```
