```python
import torch
import torch.distributions as dists
import torch.nn as nn
import torch.optim as optim
```

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
def self_normalize(log_ws):
    return nn.functional.softmax(log_ws.detach(), dim=0)
```

```python
def matmul(mat, vec):
    return torch.stack([torch.matmul(mat, vec[i]) for i in range(vec.shape[0])], dim=0)

def dot(u, v):
    return torch.stack([u[i] @ v[i] for i in range(v.shape[0])], dim=0)
```

```python
def unnormalized_log_gaussian(loc, cov, zs):
    return -1/2 * dot(zs - loc, matmul(cov.inverse(), zs - loc))
```

```python
class GaussianRWS(nn.Module):
    def __init__(self, k=100):
        super(GaussianRWS, self).__init__()
        self.k = k
        self.loc1 = nn.Parameter(torch.zeros(22))
        self.cov1 = nn.Parameter(torch.eye(22))
        
        self.loc2 = nn.Parameter(torch.ones(22) * 3)
        self.cov2 = nn.Parameter(torch.eye(22) * 3)
        
        self.register_buffer('loc3', torch.ones(22) * 1.5)
        self.register_buffer('cov3', torch.eye(22) * 0.5)
    
    def forward(self):
        objective = torch.zeros(self.k)
        
        q = dists.MultivariateNormal(self.loc1, scale_tril=torch.cholesky(self.cov1))
        zs = q.sample((self.k,))
        log_w1 = q.log_prob(zs)
        
        gamma_2 = dists.MultivariateNormal(self.loc2, scale_tril=torch.cholesky(self.cov2))
        log_gamma_2 = unnormalized_log_gaussian(gamma_2.loc, gamma_2.covariance_matrix, zs)
        log_w2 = log_gamma_2 - q.log_prob(zs)
        objective = objective + (self_normalize(log_w2) - self_normalize(log_w1)) * log_w1
        
        gamma_3 = dists.MultivariateNormal(self.loc3, scale_tril=torch.cholesky(self.cov3))
        log_gamma_3 = unnormalized_log_gaussian(gamma_3.loc, gamma_3.covariance_matrix, zs)
        log_w3 = log_w2 - log_gamma_2 + log_gamma_3
        objective = objective + (self_normalize(log_w3) - self_normalize(log_w2)) * gamma_2.log_prob(zs)
        
        kl = dists.kl.kl_divergence(gamma_3, q)
        return zs, -objective.sum(dim=0), kl
```

```python
example = GaussianRWS()
```

```python
example()
```

```python
example.train()
losses = []

optimizer = optim.Adam(example.parameters(), lr=1e-2)

for epoch in range(2000):
    optimizer.zero_grad()
    _, loss, kl = example()
    
    loss.backward()
    optimizer.step()
    losses += [kl.item()]
    print('Train epoch: {}, Inclusive KL: {:.6f}'.format(epoch, losses[-1]))
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.ylabel('RWS objective')
plt.show()
```
