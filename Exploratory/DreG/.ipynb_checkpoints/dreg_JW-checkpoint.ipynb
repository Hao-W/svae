```python
import torch as t
from torch.distributions.normal import Normal
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
t.__version__
```

```python
num_dims = 512
num_samples = 256
num_steps = 5000
learning_rate = 2e-2
```

```python
def elbo_mc(theta, phi, num_samples=num_samples):
    p = Normal(theta['loc'], theta['log_scale'].exp())
    q = Normal(phi['loc'], phi['log_scale'].exp())
    z = q.rsample((num_samples,))
    log_w = (p.log_prob(z) - q.log_prob(z)).sum(1)
    w = t.nn.functional.softmax(log_w, 0).detach()
    ess = 1.0/(w**2).sum(0)
    elbo = log_w.mean(0)
    return -elbo, ess

def elbo_iw(theta, phi, num_samples=num_samples):
    p = Normal(theta['loc'], theta['log_scale'].exp())
    q = Normal(phi['loc'], phi['log_scale'].exp())
    z = q.rsample((num_samples,))
    log_w = (p.log_prob(z) - q.log_prob(z)).sum(1)
    w = t.nn.functional.softmax(log_w, 0).detach()
    ess = 1.0/(w**2).sum(0)
    elbo = t.logsumexp(log_w, 0) - np.log(1.0 * num_samples)
    return -elbo, ess

def elbo_dr(theta, phi, num_samples=num_samples):
    p = Normal(theta['loc'], theta['log_scale'].exp())
    q = Normal(phi['loc'], phi['log_scale'].exp())
    qd = Normal(phi['loc'].detach(), phi['log_scale'].exp().detach())
    z = q.rsample((num_samples,))
    log_w = (p.log_prob(z) - qd.log_prob(z)).sum(1)
    w = t.nn.functional.softmax(log_w, 0).detach()
    ess = 1.0/(w**2).sum(0)
    elbo = (w**2 * log_w).sum(0)
    return -elbo, ess

def eubo_rw(theta, phi, num_samples=num_samples):
    p = Normal(theta['loc'], theta['log_scale'].exp())
    q = Normal(phi['loc'], phi['log_scale'].exp())
    z = q.sample((num_samples,))
    log_w = (p.log_prob(z) - q.log_prob(z)).sum(1)
    w = t.nn.functional.softmax(log_w, 0).detach()
    ess = 1.0/(w**2).sum(0)
    eubo = (w * log_w).sum(0)
    return eubo, ess

def eubo_dr(theta, phi, num_samples=num_samples):
    p = Normal(theta['loc'], theta['log_scale'].exp())
    q = Normal(phi['loc'], phi['log_scale'].exp())
    qd = Normal(phi['loc'].detach(), phi['log_scale'].exp().detach())
    z = q.rsample((num_samples,))
    log_w = (p.log_prob(z) - qd.log_prob(z)).sum(1)
    w = t.nn.functional.softmax(log_w, 0).detach()
    ess = 1.0/(w**2).sum(0)
    eubo = ((w**2 - w) * log_w).sum(0)
    return eubo,  ess


def kl(phi, theta):
    return (theta['log_scale'] 
            - phi['log_scale'] 
            + 0.5 * (phi['log_scale'].exp()**2 
                     + (phi['loc'] - theta['loc'])**2)
              / theta['log_scale'].exp()**2
            - 0.5).sum(-1)
```

```python
def train(objective, learning_rate=learning_rate, num_steps=num_steps, num_samples=num_samples):
    theta = {'loc': t.zeros(num_dims), 
             'log_scale': t.zeros(num_dims)}
    phi = {'loc': t.nn.Parameter(8.0 * t.ones(num_dims)),
           'log_scale': t.nn.Parameter(t.log(2.0 * t.ones(num_dims)))}
    output = pd.DataFrame(index=range(num_steps), columns=['objective', 'klqp', 'ess', 'loc', 'scale'])
    optimizer = t.optim.Adam(phi.values(), lr=learning_rate)
    for n in range(num_steps):
        loss, ess = objective(theta, phi, num_samples)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        output['objective'][n] = loss.detach().numpy()
        output['klqp'][n] = kl(phi, theta).detach().numpy()
        output['ess'][n] = ess.detach().numpy()
        output['loc'][n] = phi['loc'].detach().numpy()
        output['scale'][n] = np.exp(phi['log_scale'].detach().numpy())
    return output
```

```python
%time output_mc = train(elbo_mc)
```

```python
%time output_iw = train(elbo_iw)
```

```python
%time output_dr = train(elbo_dr)
```

```python
%time output_rw = train(eubo_rw)
```

```python
%time output_drw = train(eubo_dr)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12,4))

#keys = ['objective', 'klqp', 'loc', 'scale']
keys = ['objective', 'klqp', 'ess']
outputs = [output_mc, output_iw, output_dr, output_rw, output_drw]
labels = ['ELBO (MC)', 'ELBO (IW)', 'ELBO (DR)', 'EUBO (IW)', 'EUBO (DR)']

for ax, key in zip(axs.flatten(), keys):
    for output, label in zip(outputs, labels):
        if key == 'klqp':
            ax.semilogy(output[key], label=label)
        else:
            ax.plot(output[key], label=label)
    ax.set_ylabel(key)
    ax.legend()

fig.tight_layout()
```
